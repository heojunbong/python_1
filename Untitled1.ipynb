```python
from wordcloud import WordCloud
```

```python
font_path = "c:/Windows/Fonts/malgun.ttf"
```

```python
word = Okt()
```

```python
# 명사 추출하기
word_list = word.nouns(report)
word_list
```

```python
word_count = Counter(word_list)
word_count
```

```python
# 1자리 단어는 제외하고 새로 리스트 만들기
word_list_finaly = []

for w in word_list :
    if len(w) > 1 :
        word_list_finaly.append(w)
        
word_list_finaly
```

```python
word_list_finaly_count = Counter(word_list_finaly)
word_list_finaly_count
```

```python
wordcloud = WordCloud(
    # 한글폰터 설정
    font_path = font_path,
    # 배경색 설정
    background_color = "white",
    # 사용할 워드 갯수 설정
    max_words = 30,
    # 너비 설정
    width = 300,
    # 높이 설정
    height = 150
).generate_from_frequencies(word_list_finaly_count)

# 그래프 전체 크기 지정
plt.figure(figsize=(15, 10))
# 그래프 보여주기
plt.imshow(wordcloud)
# x축, y축 보여줄지 말지 설정
plt.axis("off")
# 이미지로 저장하기
plt.savefig("./report_wordcloud.png")
```
